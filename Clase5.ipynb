```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
#from sklearn.features_selection import SelectBest , f_regression
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/rama_2023_05.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x = data[['NO','PM25']]
y = data['PM10'] #variable objetivo
```

```python
split_index = int(len(data)*0.8)
x_train = x.iloc[:split_index]
x_test = x.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
svr_model.fit(x_train_scaled, y_train)
```

```python
y_test_pred = svr_model.predict(x_test_scaled)
```

```python
test_mse = mean_squared_error(y_test,y_test_pred)
test_r2= r2_score(y_test,y_test_pred)
```

```python
test_mse
```

```python
test_r2
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.iloc[split_index:]['fecha'],y_test,label='valores reales',color='green')
plt.plot(data.iloc[split_index:]['fecha'],y_test_pred,label='predicciones',color='red')
plt.xlabel('fechas')
plt.ylabel('PM10')
plt.legend()

plt.xticks(data.iloc[split_index:]['fecha'][::15],rotation=45) #[::15] marcas cada 15 dias y rotation pone las marcas en 45°
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(y_test,y_test_pred,color='blue')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--',label='linea ideal')
plt.title('valores predichos vs reales')
plt.xlabel('reales')
plt.ylabel('predicciones')
plt.legend()
plt.show()
```

### Aumento de variables temporales


```python
data['fecha'] = pd.to_datetime(data['fecha'])
```

```python
data['mes'] = data['fecha'].dt.month
```

```python
data['dia'] = data['fecha'].dt.day
```

```python
data['dia_de_semana'] = data['fecha'].dt.dayofweek
```

```python
data['fin_de_semana'] = (data['fecha'].dt.dayofweek>=5).astype(int)
# 1 sabado o domingo
# 0 luneas a viernes
```

### Seleccion de caracteristicas




```python
from sklearn.feature_selection import SelectKBest , f_regression
```

```python
X = data.drop(columns=['fecha','PM10'])
```

```python
y = data['PM10']
```

```python
selector = SelectKBest(score_func=f_regression,k=8)
x_selected = selector.fit_transform(X,y)
```

```python
x_selected
```

```python
selected_features = X.columns[selector.get_support()]
```

```python
selected_features
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)
```

```python
importancia_df = pd.DataFrame({'Caracteristica': X.columns, 'Importancia': rf_model.feature_importances_}).sort_values(by='Importancia',ascending=False)
```
